```python
from pandas import read_csv
import matplotlib.pyplot as plt

df = read_csv('dataset_1/NYC_collisions_tabular.csv')
```

```python
classes = df.PERSON_INJURY.value_counts().to_dict()

plt.title('Class distribution', pad=10.0)
plt.ylabel('Frequency', labelpad=10.0)
plt.xlabel('Class', labelpad=10.0)
plt.bar(classes.keys(), classes.values(), color=['yellow', 'red'])
plt.tight_layout()
plt.savefig("profiling/distribuition/1/class_distribuitiond1.png")
```

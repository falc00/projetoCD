```python
for i in range(len(df2.ALARM)):
    if df2.ALARM[i] == 'Safe':
        df2.ALARM[i] = 0
    else:
        df2.ALARM[i] = 1
```

```python
df2.ALARM = df2.ALARM.astype(str).astype('int64')
```

```python
corr_mtx = DataFrame(abs(df2.corr().ALARM))

fig = figure(figsize=[4, 12])

heatmap(abs(corr_mtx), xticklabels=corr_mtx.columns, yticklabels=corr_mtx.index.values, annot=True, cmap='Blues')
title('Correlation analysis')
savefig(f'images/correlation_with_class_ds2.png')
```
